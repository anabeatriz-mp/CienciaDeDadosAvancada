## Descriptive Statistics for Numerical Variables

### Report for Numerical Features Descriptive Stats 

```python
def get_descriptive_stats_numerical_features(df, numerical_features):
    
    """This function returns a table with descriptive statistics for numerical features in the DataFrame. Such as:
    - Count
    - Mean
    - Median
    - Standard Deviation
    - Coefficient of Variation
    - Minimum and Maximum
    - Quartiles (Q1 and Q3)
    - Interquartile Range (IQR)
    - Skewness
    - Kurtosis
    """

    stats_list = []

    for feature in numerical_features:

        mean = df[feature].mean()
        median = df[feature].median()
        std = df[feature].std()
        
        coef_var = std / mean if mean != 0 else np.nan
        
        min, max = df[feature].min(), df[feature].max()
        range = max - min
        q1, q3 = df[feature].quantile(0.25), df[feature].quantile(0.75)
        iqr = q3 - q1
        skewness = df[feature].skew()
        kurtosis = df[feature].kurt()

        stats_list.append({
            "Feature" : feature,
            "Count" : df[feature].count(),
            "Mean" : f"{mean:.3f}",
            "Median" : f"{median:.3f}",
            "Std" : f"{std:.3f}",
            "Coefficient of Variation" : f"{coef_var:.4f}",
            "Min" : f"{min:.3f}",
            "Max" : f"{max:.3f}",
            "range" : f"{range:.3f}",
            "Q1" : f"{q1:.3f}",
            "Q3" : f"{q3:.3f}",
            "IQR" : f"{iqr:.3f}",
            "Skewness" : f"{skewness:.4f}",
            "Kurtosis" : f"{kurtosis:.4f}"
        })
    
    return stats_list
```

```python
report = pd.DataFrame(get_descriptive_stats_numerical_features(df, numerical_features))
report
```

### Quality Table for Numerical Features Descriptive Stats

```python
def get_numerical_quality_table(report):
    
    """
    This function receives a report DataFrame and returns a table with the quality of every numerical feature based on the following criteria:
    - Coefficient of Variation (CV) >= 0.7 indicates high variability relative to the mean
    - | Skewness | > 1  indicates a highly skewed distribution 
    - | Skewness | >= 0.5 indicates a moderately skewed distribution
    - Kurtosis >= 3 indicates heavy tails
    - Kurtosis high extreme values (e.g., > 10) indicate the presence of strong outliers
    """

    quality_table = []

    for _ , row in report.iterrows():

        quality_list = []

        feature = row['Feature']
        cv = pd.to_numeric(row['Coefficient of Variation'], errors='coerce')
        skewness = pd.to_numeric(row['Skewness'], errors='coerce')
        kurtosis = pd.to_numeric(row['Kurtosis'], errors='coerce')

        if cv >= 0.7:
            quality_list.append('High Variability')
        
        if abs(skewness) > 1:
            quality_list.append('Highly Skewed')
        elif abs(skewness) >= 0.5:
            quality_list.append('Moderately Skewed')


        if kurtosis >= 3 and kurtosis <= 10:
            quality_list.append('Heavy Tails')
        elif kurtosis > 10:
            quality_list.append('Extreme Outliers')

        quality_table.append({
            "Feature": feature,
            "Challenges": ', '.join(quality_list) if quality_list else 'Statistically well-behaved'
        })

    return pd.DataFrame(quality_table)
```

```python
quality_report = get_numerical_quality_table(report)
quality_report
```

## Univariate Analysis of Categorical Variables

```python
def create_report_categorical_features(df, categorical_features, rare_threshold=0.05):
    """
    This function receives a DataFrame and a list of categorical features and returns a report with 
    the following information for each categorical feature:
    - Absolute frequency of each category
    - Cardinality (number of unique categories)
    - Cardinality Level (Low, Medium, High)
    - Biggest Category Percentage
    - Number of Rare Levels (categories with low frequency, e.g., less than 5% of the total, or with less than 10 occurrences)
    - Percentage of Rare Levels
    - Missing Percentage
    """

    report = []

    for col in categorical_features:
       
        counts = df[col].value_counts()
        rel_freqs = df[col].value_counts(normalize=True)
        
        total_observations = counts.sum()
        cardinality = len(counts)
        biggest_category_pct = rel_freqs.max() * 100

        rare_mask = (rel_freqs < rare_threshold) | (counts < 10)
        rare_count = rare_mask.sum()
        rare_pct = rel_freqs[rare_mask].sum() * 100

        missing_pct = df[col].isnull().mean() * 100

        def cardinality_level(cardinality):
            if cardinality < 10:
                return 'Low Cardinality'
            elif cardinality <= 30:
                return 'Medium Cardinality'
            else:
                return 'High Cardinality'

        report.append({
            "Feature": col,
            "Total Non-Nulls": total_observations,
            "Cardinality": cardinality,
            "Cardinality Level": cardinality_level(cardinality),
            "Biggest Category %": f"{biggest_category_pct:.2f}%",
            "Rare Levels Count": rare_count,
            "Rare Levels Total %": f"{rare_pct:.2f}%",
            "Missing %": f"{missing_pct:.2f}%"
        })

    return pd.DataFrame(report)
```

```python
categorical_report = create_report_categorical_features(df, categorical_features)
categorical_report
```
